### 53. Maximum Subarray
#### Given an integer array nums, find the subarray with the largest sum, and return its sum.
#### Example: Input: nums = [-2,1,-3,4,-1,2,1,-5,4], Output: 6

### Description: Kadane's Algorithm
#### Kadane's Algorithm is an iterative dynamic programming technique used to find the maximum sum of a contiguous subarray within a one-dimensional array of numbers.
#### time complexity of O(n) and a space complexity of O(1).
### When to Identify the Need for Kadane's Algorithm:
#### Contiguous Subarrays: The problem explicitly asks for a contiguous sequence, segment, or sub-array.
#### Maximum or Minimum Optimization: The goal is to maximize or minimize a specific aggregated value.
#### Mixed Values: The array contains a mix of positive and negative numbers. If an array has only positive numbers, the maximum subarray is simply the entire array. The negative numbers are what create the challenge.

### Java Solution
```java
class Solution {
    public int maxSubArray(int[] nums) {
        int current_sum = 0;
        int max_sum = nums[0];
        for(int i=0;i<nums.length;i++){
            current_sum += nums[i];
            if(current_sum > max_sum){
                max_sum = current_sum;
            }
            if(current_sum < 0){
                current_sum = 0;
            }
        }
        return max_sum;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    int maxSubArray(vector<int>& nums) {
        // Initialize variables to the first element
        int current_sum = nums[0];
        int max_sum = nums[0];
        
        // Loop through the array starting from index 1
        for (int i = 1; i < nums.size(); i++) {
            // The core Kadane's decision using std::max
            current_sum = max(nums[i], current_sum + nums[i]);
            
            // Update the global maximum if necessary
            max_sum = max(max_sum, current_sum);
        }
        
        return max_sum;
    }
};
```

### Python Solution

````python
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        # Initialize variables with the first element of the array
        current_sum = max_sum = nums[0]
        
        # Iterate through the array starting from the second element
        for num in nums[1:]:
            # The core Kadane's decision
            current_sum = max(num, current_sum + num)
            
            # Update the global maximum if necessary
            max_sum = max(max_sum, current_sum)
            
        return max_sum
````
