### 912. Sort an Array
### Given an array of integers nums, sort the array in ascending order and return it.
### Example: Input: nums = [5,1,1,2,0,0]
#### Output: [0,0,1,1,2,5]

### Java Solution
```java
class Solution {
    public int[] sortArray(int[] nums) {
        Map<Integer,Integer> map = new HashMap<>();
        int max = nums[0];
        int min = nums[0];
        for(int i=0;i<nums.length;i++){
            if(map.containsKey(nums[i])){
                map.put(nums[i], map.get(nums[i]) + 1);
            }
            else{
                map.put(nums[i],1);
            }
            if(nums[i]<min){
                min = nums[i];
            }
            if(nums[i]>max){
                max = nums[i];
            }
        }
        int index = 0;
        for(int i=min;i<=max;i++){
            while(map.getOrDefault(i,0)>0){
                nums[index] = i;
                index++;
                map.put(i,map.getOrDefault(i, 0) - 1);
            }
        }
        return nums;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    vector<int> sortArray(vector<int>& nums) {
        unordered_map<int, int> countMap;
        int max_val = nums[0];
        int min_val = nums[0];
        
        for (int i = 0; i < nums.size(); i++) {
            // In C++, this automatically handles initialization to 0 if the key doesn't exist
            countMap[nums[i]]++;
            
            if (nums[i] < min_val) {
                min_val = nums[i];
            }
            if (nums[i] > max_val) {
                max_val = nums[i];
            }
        }
        
        int index = 0;
        for (int i = min_val; i <= max_val; i++) {
            while (countMap[i] > 0) {
                nums[index] = i;
                index++;
                countMap[i]--;
            }
        }
        
        return nums;
    }
};
```

### Python Solution

````python
class Solution:
    def sortArray(self, nums):
        count_map = {}
        max_val = nums[0]
        min_val = nums[0]
        
        for num in nums:
            # Equivalent to Java's map.put(nums[i], map.getOrDefault(...) + 1)
            count_map[num] = count_map.get(num, 0) + 1
            
            if num < min_val:
                min_val = num
            if num > max_val:
                max_val = num
                
        index = 0
        for i in range(min_val, max_val + 1):
            while count_map.get(i, 0) > 0:
                nums[index] = i
                index += 1
                count_map[i] -= 1
                
        return nums
````
