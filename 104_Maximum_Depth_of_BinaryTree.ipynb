### 104. Maximum Depth of Binary Tree
#### Given the root of a binary tree, return its maximum depth.
#### A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.
#### Example: 

### Description: 
#### Base Case: If the tree is empty (the node is NULL), its depth is 0.
#### Recursive Step: For any given node, find the depth of its left subtree and the depth of its right subtree.
#### Combine: Take the maximum of those two depths and add 1 (to account for the current node itself).

### Java Solution
```java
public class TreeDepth {
    public int maxDepth(TreeNode root) {
        // Base case: if we hit a null node, the depth path ends
        if (root == null) {
            return 0;
        }

        // Recursively calculate the depth of left and right subtrees
        int leftDepth = maxDepth(root.left);
        int rightDepth = maxDepth(root.right);

        // Return the greater of the two depths, plus 1 for the current level
        return Math.max(leftDepth, rightDepth) + 1;
    }
}
```

### Cpp Solution
```cpp
int maxDepth(TreeNode* root) {
    // Base case: if we hit a null node, the depth path ends
    if (root == NULL) {
        return 0;
    }

    // Recursively calculate the depth of left and right subtrees
    int leftDepth = maxDepth(root->left);
    int rightDepth = maxDepth(root->right);

    // Return the greater of the two depths, plus 1 for the current level
    return std::max(leftDepth, rightDepth) + 1;
}
```

### Python Solution

````python
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        # Base case: if the node is None, depth is 0
        if not root:
            return 0
        
        # Recursively calculate left and right depths
        # Note the use of 'self.' to call the class method
        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        
        # Return the max of the two sides + 1 for the current node
        return max(left_depth, right_depth) + 1
````
