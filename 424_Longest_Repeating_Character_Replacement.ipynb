### 424. Longest Repeating Character Replacement
#### You are given a string s and an integer k. You can choose any character of the string and change it to any other uppercase English character. You can perform this operation at most k times.
#### Return the length of the longest substring containing the same letter you can get after performing the above operations.
#### Example: Input: s = "ABAB", k = 2
#### Output: 4

### Java Solution
```java
class Solution {
    public int characterReplacement(String s, int k) {
        int[] count = new int[26];
        int left = 0;
        int max_length =0;
        int max_count = 0;
        for(int right=0;right<s.length();right++){
            char ch = s.charAt(right);
            count[ch-'A']++;
            max_count = Math.max(max_count,count[ch-'A']);

            while (right - left + 1 - max_count > k) {
                char leftChar = s.charAt(left);
                count[leftChar - 'A']--;
                left++;
            }
            max_length = Math.max(max_length, right - left + 1);
        }
        return max_length;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    int characterReplacement(string s, int k) {
        vector<int> charCounts(26, 0);
        int left = 0;
        int maxLength = 0;
        int maxCount = 0;
        
        for (int right = 0; right < s.length(); right++) {
            // Increment the frequency of the current character
            charCounts[s[right] - 'A']++;
            
            // Update maxCount with the highest frequency in the window
            maxCount = max(maxCount, charCounts[s[right] - 'A']);
            
            // If the window is invalid, slide the left pointer forward
            while ((right - left + 1) - maxCount > k) {
                charCounts[s[left] - 'A']--;
                left++;
            }
            
            // Update maxLength found so far
            maxLength = max(maxLength, right - left + 1);
        }
        
        return maxLength;
    }
};
```

### Python solution

````python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        char_counts = [0] * 26
        left = 0
        max_length = 0
        max_count = 0
        
        for right in range(len(s)):
            # Convert char to 0-25 index using ord()
            idx = ord(s[right]) - ord('A')
            char_counts[idx] += 1
            
            # Update the highest frequency found in the current window
            max_count = max(max_count, char_counts[idx])
            
            # If remaining characters exceed k, shrink the window from the left
            while (right - left + 1) - max_count > k:
                left_idx = ord(s[left]) - ord('A')
                char_counts[left_idx] -= 1
                left += 1
                
            # Update overall max length
            max_length = max(max_length, right - left + 1)
            
        return max_length
````
