### 141. Linked List Cycle
#### Given head, the head of a linked list, determine if the linked list has a cycle in it.
#### There is a cycle in a linked list if there is some node in the list that can be reached again by continuously following the next pointer. Internally, pos is used to denote the index of the node that tail's next pointer is connected to. Note that pos is not passed as a parameter.
#### Return true if there is a cycle in the linked list. Otherwise, return false.
#### Example: Input: head = [3,2,0,-4], pos = 1, Output: true
#### Explanation: There is a cycle in the linked list, where the tail connects to the 1st node (0-indexed).

### Description:
#### The most elegant way to detect a cycle in a linked list relies on a concept called Floyd’s Cycle-Finding Algorithm, often referred to as the "Tortoise and Hare" approach.
#### Imagine two runners starting at the exact same point.
#### The Slow Runner (Tortoise): Takes one step at a time.
#### The Fast Runner (Hare): Takes two steps at a time.
#### If the track is a straight line (a normal linked list), the fast runner will simply hit the finish line and stop.
#### But what if the track has a loop in it? The fast runner will never hit a finish line. Instead, they will run into the loop, and eventually, the slow runner will enter the loop too. Because the fast runner is moving quicker, they will continuously circle the loop and eventually lap the slow runner from behind.
#### The core logic: If the fast runner and the slow runner ever land on the exact same spot at the exact same time, the track must be a loop. 

### Java Solution
```java
public class Solution {
    public boolean hasCycle(ListNode head) {
        ListNode slow = head;
        ListNode fast = head;
        while(fast!=null && fast.next!=null){
            slow = slow.next;
            fast = fast.next.next;
            if(slow==fast){
                return true;
            }
        }
        return false;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    bool hasCycle(ListNode *head) {
        ListNode* slow = head;
        ListNode* fast = head;
        
        while (fast != nullptr && fast->next != nullptr) {
            slow = slow->next;
            fast = fast->next->next;
            
            if (slow == fast) {
                return true;
            }
        }
        
        return false;
    }
};
```
