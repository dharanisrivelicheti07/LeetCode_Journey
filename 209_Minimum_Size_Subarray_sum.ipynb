### 209. Minimum Size Subarray Sum
#### Given an array of positive integers nums and a positive integer target, return the minimal length of a subarray whose sum is greater than or equal to target. If there is no such subarray, return 0 instead.
#### Example: Input: target = 7, nums = [2,3,1,2,4,3]
#### Output: 2

### Description
#### Sliding window is an optimization technique used for arrays and substrings to solve problems like subarrays and substrings. It is very useful for Continguous Elements. It reduces the time complexity from O(n^2) to O(n).
#### Brute Force approach recalculates every subarray / substring from scratch. For each starting index, it scans again up to k elements. This leads to many repeated additions and usually takes O(n²). Sliding Window reuses previous work by removing the element that goes out and adding the new element that comes in. Thus, each step does O(1) work and the whole process becomes O(n)

### Java Solution
```java
lass Solution {
    public int minSubArrayLen(int target, int[] nums) {
        int left = 0;
        int right = 0;
        int currentsum = 0;
        int min_length = Integer.MAX_VALUE;
        while(right<nums.length){
            currentsum += nums[right];
            while(currentsum>=target){
                min_length=Math.min(min_length,right-left+1);
                currentsum -= nums[left];
                left++;
            }
            right++;
        }
        return min_length == Integer.MAX_VALUE ? 0 : min_length;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    int minSubArrayLen(int target, vector<int>& nums) {
        // 1. Initialize pointers
        int left = 0;
        int right = 0;
        
        // 2. Initialize state variables to track what's inside the window
        int currentSum = 0;
        
        // 3. Initialize the result variable 
        // Use INT_MAX to represent infinity for finding the minimum
        int minLength = INT_MAX;

        // 4. Start expanding the window using the right pointer
        while (right < nums.size()) {
            
            // STEP A: Add the current element on the right to our window state
            currentSum += nums[right];

            // STEP B: Check if the window is VALID (sum >= target)
            while (currentSum >= target) {
                
                // Track the minimum length found so far
                minLength = min(minLength, right - left + 1);
                
                // STEP C: Remove the left element from your window state
                currentSum -= nums[left];
                
                // STEP D: Shrink the window
                left++;
            }

            // STEP E: Move the right pointer forward to keep expanding
            right++;
        }

        // If minLength was never updated, return 0
        return minLength == INT_MAX ? 0 : minLength;
    }
};
```

### Python solution

````python
class Solution:
    def minSubArrayLen(self, target: int, nums: list[int]) -> int:
        # 1. Initialize the left pointer and state variable
        left = 0
        current_sum = 0
        
        # 3. Initialize the result variable using Python's float infinity
        min_length = float('inf')

        # 4. Start expanding the window 
        # Using a for-loop naturally handles the 'right' pointer incrementing
        for right in range(len(nums)):
            
            # STEP A: Add the current element to our window state
            current_sum += nums[right]

            # STEP B: Check if the window is VALID (sum >= target)
            while current_sum >= target:
                
                # Track the minimum length found so far
                min_length = min(min_length, right - left + 1)
                
                # STEP C: Remove the left element from your window state
                current_sum -= nums[left]
                
                # STEP D: Shrink the window
                left += 1

        # If min_length was never updated, return 0
        return min_length if min_length != float('inf') else 0
````
