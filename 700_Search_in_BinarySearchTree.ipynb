### 700. Search in a Binary Search Tree
#### You are given the root of a binary search tree (BST) and an integer val.
#### Find the node in the BST that the node's value equals val and return the subtree rooted with that node. If such a node does not exist, return null.

### Description
#### In a binary search tree the nodes are inserted in the order of which the root node is less than or greater.If the value is less than the root node then it is inserted in left side and if the value is greater than the root node then the node is inserted right side of the tree. Left Subtree: All nodes in the left subtree must have a value strictly less than the node's value. Right Subtree: All nodes in the right subtree must have a value strictly greater than the node's value.Search, Insert, and Delete (Average Case): O(log n). Search, Insert, and Delete (Worst Case): O(n)

### Java Solution
```java
class Solution {
    public TreeNode searchBST(TreeNode root, int val) {
       while(root != null && root.val != val){
        if(val < root.val){
            root = root.left;
        }
        else{
            root = root.right;
        }
       }
       return root;
    }
        
    }
```    

### Cpp Solution
```cpp
class Solution {
public:
    TreeNode* searchBST(TreeNode* root, int val) {
        // Traverse until we hit a leaf or find the target
        while (root != nullptr && root->val != val) {
            if (val < root->val) {
                root = root->left;
            } else {
                root = root->right;
            }
        }
        return root;
    }
};
```

### Python Solution

````python
class Solution:
    def searchBST(self, root: TreeNode, val: int) -> TreeNode:
        # Traverse down the tree
        while root and root.val != val:
            if val < root.val:
                root = root.left
            else:
                root = root.right
                
        # Returns the node if found, or None if the loop finished without a match
        return root
````
