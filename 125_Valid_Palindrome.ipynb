### 125. Valid Palindrome
#### A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers.
#### Given a string s, return true if it is a palindrome, or false otherwise.
#### Example: Input: s = "A man, a plan, a canal: Panama", Output: true

### Java Solution
```java
class Solution {
    public static boolean checkPalindrome(String s){
        int left = 0;
        int right = s.length()-1;
        while(left<right){
            if(s.charAt(left)!=s.charAt(right)){
                return false;
            }
            left++;
            right--;
        }
        return true;
    }
    public boolean isPalindrome(String s) {
        StringBuilder sb = new StringBuilder();
        for(int i=0;i<s.length();i++){
            char ch = s.charAt(i);
            if(Character.isLetter(ch) || Character.isDigit(ch)){
                sb.append(ch);
            }
        }
        String result = sb.toString();
        result = result.toLowerCase();
        return checkPalindrome(result);
    }
}
```

### Cpp Solution
```cpp
class Solution {
private:
    // Helper function equivalent to your static checkPalindrome
    bool checkPalindrome(const std::string& s) {
        int left = 0;
        int right = s.length() - 1;
        while (left < right) {
            if (s[left] != s[right]) {
                return false;
            }
            left++;
            right--;
        }
        return true;
    }

public:
    bool isPalindrome(std::string s) {
        std::string result = "";
        
        // Equivalent to your StringBuilder loop
        for (char ch : s) {
            if (std::isalnum(ch)) {
                result += std::tolower(ch);
            }
        }
        
        return checkPalindrome(result);
    }
};
```

### Python Solution


````python
class Solution:
    # Helper method
    def checkPalindrome(self, s: str) -> bool:
        left = 0
        right = len(s) - 1
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True

    def isPalindrome(self, s: str) -> bool:
        # Equivalent to your StringBuilder logic
        sb = []
        for ch in s:
            if ch.isalnum():
                sb.append(ch.lower())
        
        # Convert the list of characters back into a string
        result = "".join(sb)
        
        return self.checkPalindrome(result)
````
