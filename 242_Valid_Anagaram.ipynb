### 242. Valid Anagram
#### Given two strings s and t, return true if t is an anagram of s, and false otherwise.
#### Example: Input: s = "anagram", t = "nagaram", Output: true
#### Example: Input: s = "rat", t = "car", Output: false


### Description:
#### An anagram is a word or phrase formed by rearranging the letters of a different word or phrase. To be a true anagram, the new word or phrase must use all the original letters exactly once.
#### Time Complexity:O(n), Space Complexity:O(1).
#### The Sentence ch-'a' represents: 
#### If the letter is 'a': 'a' - 'a' translates to 97 - 97 = 0 (Maps to array index 0)
#### If the letter is 'b': 'b' - 'a' translates to 98 - 97 = 1 (Maps to array index 1)
#### If the letter is 'c': 'c' - 'a' translates to 99 - 97 = 2 (Maps to array index 2)
#### If the letter is 'z': 'z' - 'a' translates to 122 - 97 = 25 (Maps to array index 25)

### Java Solution
```java
class Solution {
    public boolean isAnagram(String s, String t) {
        if(s.length() != t.length()){
            return false;
        }
        int count[] = new int[26];
        for(int i=0;i<s.length();i++){
            char ch = s.charAt(i);
            count[ch-'a']++;
        }
        for(int i=0;i<t.length();i++){
            char ch = t.charAt(i);
            count[ch-'a']--;
        }
        for(int i=0;i<count.length;i++){
            if(count[i]>0){
                return false;
            }
        }
        return true;
    }
}
```

### Cpp Solution
```cpp
class Solution {
public:
    bool isAnagram(string s, string t) {
        // If lengths differ, they cannot be anagrams
        if (s.length() != t.length()) {
            return false;
        }

        // Initialize a frequency array for the 26 lowercase letters with 0s
        int counts[26] = {0};

        // Increment for s, decrement for t
        for (int i = 0; i < s.length(); i++) {
            counts[s[i] - 'a']++;
            counts[t[i] - 'a']--;
        }

        // Check if all counts are balanced back to 0
        for (int count : counts) {
            if (count != 0) {
                return false;
            }
        }

        return true;
    }
};
```

### Python Solution

````python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
            
        # Create an array of 26 zeros
        counts = [0] * 26
        
        # Traverse strings and update frequencies
        for i in range(len(s)):
            counts[ord(s[i]) - ord('a')] += 1
            counts[ord(t[i]) - ord('a')] -= 1
            
        # Check for any non-zero counts
        for count in counts:
            if count != 0:
                return False
                
        return True
````
