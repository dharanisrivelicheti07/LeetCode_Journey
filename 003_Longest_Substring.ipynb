### 3. Longest Substring Without Repeating Characters


#### Given a string s, find the length of the longest substring without duplicate characters.
#### Example 1:
#### Input: s = "abcabcbb"
#### Output: 3
#### Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

### Description

##### HashSet is a set which doesnot contain duplicate elements it only contains keys it doesnt have values.It is under two packages called java.util.HashSet and java.util.Set and we need to import that when we solve any problem under the topic of hashset and it is represented as same as hashmap and c++ it is known as unordered_set and in python it is called set(). There are several functions they are add() which adds element to the set, and remove() which removes or deletes the element from the set, and contains() which says that checks if an element exists in the set or not.Sliding window is the technique we used here where we used two pointers called left and right which points to the starting index of the string and right pointer is fast forward and left pointer moves only when it violates any rule. the length is calculated using right-left+1 amd the maximum is calculated using the Math.max() function this max value represent the longest substring.

### Java Solution
```java
class Solution {
    public int lengthOfLongestSubstring(String s) {
        Set<Character> set = new HashSet<>();
        int max = 0;
        int left = 0;
        for(int right = 0;right<s.length();right++){
            char ch = s.charAt(right);
            while(set.contains(ch)){
                set.remove(s.charAt(left));
                left++;
            }
            set.add(ch);
            int len = right-left+1;
            max = Math.max(max,len);
        }
        return max;
    }
}
```

### Cpp Solution
```cpp
#include <string>
#include <unordered_set>
#include <algorithm> // For std::max

using namespace std;

class Solution {
public:
    int lengthOfLongestSubstring(string s) {
        // In C++, the equivalent of HashSet is unordered_set
        unordered_set<char> set;
        int max_len = 0;
        int left = 0;
        
        for(int right = 0; right < s.length(); right++) {
            char ch = s[right];
            
            // .count() returns 1 if the element exists, 0 otherwise. 
            // (In C++20, you can also use .contains(ch))
            while(set.count(ch)) {
                set.erase(s[left]); // Remove the character at the left pointer
                left++;             // Shrink the window
            }
            
            set.insert(ch);         // Add the current character
            int len = right - left + 1;
            max_len = max(max_len, len); // Update max length
        }
        
        return max_len;
    }
};
```

### Python Solution

````python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        # In Python, a hash set is simply initialized with set()
        char_set = set()
        max_len = 0
        left = 0
        
        # 'right' automatically increments in the for loop
        for right in range(len(s)):
            ch = s[right]
            
            # The 'in' keyword is used to check for existence in the set
            while ch in char_set:
                char_set.remove(s[left]) # Remove the character at the left pointer
                left += 1                # Shrink the window
                
            char_set.add(ch)             # Add the current character
            length = right - left + 1
            max_len = max(max_len, length) # Update max length
            
        return max_len
````
