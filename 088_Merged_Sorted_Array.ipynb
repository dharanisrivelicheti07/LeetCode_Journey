### 88. Merge Sorted Array
#### You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.
#### Merge nums1 and nums2 into a single array sorted in non-decreasing order.
#### The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

#### Example: Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
#### Output: [1,2,2,3,5,6]
#### Explanation: The arrays we are merging are [1,2,3] and [2,5,6].The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1

### Description:
#### The two pointer concept is one of the most powerful optimization technique. It involves using two variables to traverse a data structure simulatenously. One of the method of two pointer concept is "Pointer on two different arrays" which means when we have two seperate lists of data placing one pointer at the start of the first array and the second pointer at the start of the second array.

#### Difference Between Two Pointers and Sliding Window:


#### Logic for Pointers on Two Different Arrays:

````python
def merge_sorted_arrays(arr1, arr2):
    i = 0  # Pointer for arr1
    j = 0  # Pointer for arr2
    merged = []
    
    # Traverse both arrays as long as BOTH have elements left
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            merged.append(arr1[i])
            i += 1  # Move arr1 pointer forward
        else:
            merged.append(arr2[j])
            j += 1  # Move arr2 pointer forward
            
    # If there are leftovers in arr1, add them
    while i < len(arr1):
        merged.append(arr1[i])
        i += 1
        
    # If there are leftovers in arr2, add them
    while j < len(arr2):
        merged.append(arr2[j])
        j += 1
        
    return merged
````

class Solution {
    public void merge(int[] nums1, int m, int[] nums2, int n) {
        int[] temp= new int[m+n];
        int i = 0;
        int j = 0;
        int k = 0;
        while(i<m&&j<n){
            if(nums1[i]<nums2[j]){
                temp[k] = nums1[i];
                i++;
            }
            else{
                temp[k]=nums2[j];
                j++;
            }
            k++;
        }
        while(i<m){
            temp[k]=nums1[i];
            i++;
            k++;
        }
        while(j<n){
            temp[k]=nums2[j];
            j++;
            k++;
        }
        for(int x = 0;x<m+n;x++){
            nums1[x]=temp[x];
        }
    }
}

class TwoArrays {
public:
    vector<int> mergeSortedArrays(vector<int>& arr1, vector<int>& arr2) {
        // Pre-allocating the exact size needed (highly recommended in C++ for memory efficiency)
        vector<int> merged(arr1.size() + arr2.size());
        
        int i = 0; // Pointer for arr1
        int j = 0; // Pointer for arr2
        int k = 0; // Pointer for the merged array
        
        // Traverse both arrays as long as BOTH have elements left
        while (i < arr1.size() && j < arr2.size()) {
            if (arr1[i] < arr2[j]) {
                merged[k] = arr1[i];
                i++;
            } else {
                merged[k] = arr2[j];
                j++;
            }
            k++;  // Move merged array pointer forward
        }
        
        // If there are leftovers in arr1, add them
        while (i < arr1.size()) {
            merged[k] = arr1[i];
            i++;
            k++;
        }
        
        // If there are leftovers in arr2, add them
        while (j < arr2.size()) {
            merged[k] = arr2[j];
            j++;
            k++;
        }
        
        return merged;
    }
};
