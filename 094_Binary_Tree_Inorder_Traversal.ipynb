### 94. Binary Tree Inorder Traversal
#### Given the root of a binary tree, return the inorder traversal of its nodes' values.
#### Example 1: Input: root = [1,null,2,3]
#### Output: [1,3,2]

### Description
#### Binary Tree is a hierarchical data structure it is different from arrays and linkedlist they are linear which means one after the other,it is a tree branches out. Binary means two which has two childs means the left child and the right child and top most node is called root node and for every tree it has only on root node and the leaf nodes means they have no other child nodes it refers to the case which is a null values. The Inorder traversal of the binarytree is Left->Root->Right.

### Java Solution
```java
class Solution {
    public List<Integer> inorderTraversal(TreeNode root) {
        List<Integer> result = new ArrayList<>();
        traverse(root, result);
        return result;
    }
    public void traverse(TreeNode node,List<Integer> result){
        if(node==null){
            return;
        }
        traverse(node.left,result);
        result.add(node.val);
        traverse(node.right,result);
    }
}
```

### Description for java solution
#### public List<Integer> inorderTraversal(TreeNode root): This is the main function signature. It takes the root (the very top node) of the tree and promises to return a List of integers.
#### List<Integer> result = new ArrayList<>();	We create an empty, dynamic array called result. This will store the values of our nodes as we visit them in the correct order.
#### traverse(root, result); We call our private helper function, passing it the starting root node and our empty result list so it can be filled.
#### return result;	Once the traverse function finishes, the result list is full. We return it to the caller.
#### private void traverse(TreeNode node, List<Integer> result) A helper function. It needs to know which node it is currently looking at, and it needs access to the result list to add values to it.
##### if (node == null) { return; }	The Base Case: If we hit a dead end (a non-existent node), we stop and go back. This prevents infinite loops and crashes.
#### traverse(node.left, result);	Step 1 of Inorder: Recursively travel down the entirely of the left subtree first.
#### result.add(node.val); Step 2 of Inorder: Process the current node. We extract its value (node.val) and append it to our list.
#### traverse(node.right, result);	Step 3 of Inorder: Recursively travel down the entirely of the right subtree last.

### Cpp Solution
```cpp
class Solution {
public:
    vector<int> inorderTraversal(TreeNode* root) {
        vector<int> result;
        traverse(root, result);
        return result;
    }

private:
    void traverse(TreeNode* node, vector<int>& result) {
        if (node == nullptr) {
            return;
        }
        traverse(node->left, result);
        result.push_back(node->val);
        traverse(node->right, result);
    }
};
```

### Description for cpp solution
#### vector<int> inorderTraversal(TreeNode* root)	The main function. It takes a pointer to the root node and returns a vector (a resizable array) of integers.
#### vector<int> result;	We create an empty vector named result to store our values.
#### traverse(root, result); We call our helper function. Notice we are passing the result vector directly to it.
#### void traverse(TreeNode* node, vector<int>& result)	The helper function. The & symbol is crucial here. It means we are passing result by reference. We are giving the function the actual vector, not a copy of it, so changes made here affect the original.
#### if (node == nullptr) { return; }	Base Case: If the current pointer doesn't point to a valid node (nullptr is C++'s version of null), stop and go back.
#### traverse(node->left, result);	Step 1 (Left): Because node is a pointer, we use the arrow operator -> to access its left child, and recursively traverse it.
#### result.push_back(node->val);	Step 2 (Root): We access the node's value (node->val) and add it to the end of our vector using .push_back().
#### traverse(node->right, result);	Step 3 (Right): We use the arrow operator to access the right child and recursively traverse it.

### Python Solution

````python
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        result = []
        
        def traverse(node):
            if not node:
                return
            
            traverse(node.left)
            result.append(node.val)
            traverse(node.right)
            
        traverse(root)
        return result
````

### Description for Python Solution
#### def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:	The main method. Optional[TreeNode] means the root can be a TreeNode or None. -> List[int] means it promises to return a list of integers. self is required because this function belongs to a class.
#### result = [] We initialize result as an empty list (Python's built-in dynamic array).
#### def traverse(node): We define a nested helper function inside our main function. The beauty of this is that traverse can automatically see and modify the result list we created above. We don't have to pass it as an argument!
#### if not node: Base Case: This is Python's elegant way of saying if node == None:. If the node doesn't exist, stop and return.
#### traverse(node.left) Step 1 (Left): Recursively call the function on the left child.
#### result.append(node.val) Step 2 (Root): Extract the value and add it to our result list using .append().
#### traverse(node.right) Step 3 (Right): Recursively call the function on the right child.
#### traverse(root)	Down here, we actually kick off the recursion by calling the nested helper function on our starting root.
#### return result	Once the recursion finishes, the list is fully populated, and we return it.
